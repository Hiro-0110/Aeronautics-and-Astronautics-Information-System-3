```python
learning_rate = 0.0005
iteration = 1000000
history = {'epsilon_squared':[], 'T0_diff':[], 'a_diff':[], 'omega_diff':[], 'theta_diff':[]}

T0 = 12.7
a = 3
theta = 0
omega = 2 * np.pi / 6000


for number in range(iteration):

  epsilon_squared = 0
  T0_diff = 0
  a_diff = 0
  omega_diff = 0
  theta_diff = 0
  epsilon_list = []

  for i in range(len(df)):
    T0_diff += -2 * ((df.iloc[i, 1]) - T0 - a * np.sin(omega * df.iloc[i, 0] + theta))
    a_diff +=  -2 * np.sin(omega * df.iloc[i, 0] + theta) * ((df.iloc[i, 1]) - T0 - a * np.sin(omega * df.iloc[i, 0] + theta) )
    omega_diff += -2 * a  * np.cos(omega*df.iloc[i, 0] + theta) * ((df.iloc[i, 1]) - T0 - a * np.sin(omega * df.iloc[i, 0] + theta) )
    theta_diff += -2 * a * np.cos(omega*df.iloc[i, 0] + theta) * ((df.iloc[i, 1]) - T0 - a * np.sin(omega * df.iloc[i, 0] + theta) )
    epsilon_squared += ((df.iloc[i, 1]) - T0 - a * np.sin(omega * df.iloc[i, 0] + theta) )**2

  sigma_squared = epsilon_squared/len(df)

  history['epsilon_squared'].append(epsilon_squared)
  history['T0_diff'].append(T0_diff)
  history['a_diff'].append(a_diff)
  history['omega_diff'].append(omega_diff)
  history['theta_diff'].append(theta_diff)

  T0 -= T0_diff * learning_rate
  a -= a_diff * learning_rate/100
  omega -= omega_diff * learning_rate/1000
  theta -= theta_diff * learning_rate

  if ((number+1) % 5000 == 0) or (number == 0) :
    print(f"--------------{number+1}回目---------------------")
    print(f"T0:{T0}, a:{a}, omega:{omega}, theta:{theta}, sigma_squared:{sigma_squared}")
    print(f"T0_diff:{T0_diff}, a_diff:{a_diff}, omega_diff:{omega_diff}, theta_diff:{theta_diff}")
    #print(f"epsilon_squared:{epsilon_squared}")
    for i in range(len(df)):
      epsilon_list.append(np.random.normal(0, np.sqrt(sigma_squared)))
    plt.scatter(df['Time'],df['Temp'], s=10)
    plt.plot(df['Time'],T0+a*np.sin(omega*df['Time']+theta)+epsilon_list, color='orange')
    plt.xlabel("Time")
    plt.ylabel("Temperature")
    plt.title(f"iteration : {number+1} , epsilon_squared : {epsilon_squared}")
    plt.savefig(f"graphs/{number+1}回目.png", dpi=300)
    plt.show()
  if T0_diff<1e-04 and a_diff<1e-04 and omega_diff<1e-04 and theta_diff < 1e-04:
    break
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.model_selection import train_test_split
import numpy.linalg as la
```

```python
df = pd.read_csv('/content/mazdas.txt', sep='\t')
```

```python
df
```

```python
plt.scatter(df['Year'], df['Price'], s=15)
plt.xlabel('Year')
plt.ylabel('Price')
plt.savefig("fig1.png", dpi=300)
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(df['Year'].values, df['Price'].values, test_size=0.3, random_state=0)
```

```python
X_train_p = np.concatenate([np.ones((len(X_train), 1)), X_train.reshape((len(X_train), 1)) ], 1)
```

```python
alp = la.solve(np.dot(X_train_p.T, X_train_p), np.dot(X_train_p.T, y_train))
print('alp = ',alp)
#Predicted Prices
yhat = np.dot(X_train_p,alp)
#Sum of squared errors (rediduals)
Q = np.sum((y_train-yhat)**2)
print('Q = %.2e' %Q)
```

```python
fig = plt.figure(figsize=(9,6))
#training data
ax1 = fig.add_subplot(1,2,1)
ax1.scatter(X_train, y_train, s=15)
ax1.set_xlabel('Year')
ax1.set_ylabel('Price')
ax1.plot(X_train, yhat, color='orange')
ax1.set_title(f'train : Linear Regression, Q : {Q:.3f}')
#test data
ax2 = fig.add_subplot(1,2,2)
ax2.scatter(X_test, y_test, s=15)
ax2.set_xlabel('Year')
ax2.set_ylabel('Price')
X_test_p = np.concatenate([np.ones((len(X_test), 1)), X_test.reshape((len(X_test), 1)) ], 1)
yhat_test = np.dot(X_test_p,alp)
Q_test = np.sum((y_test-yhat_test)**2)
ax2.plot(X_test, yhat_test, color='orange')
ax2.set_title(f'test : Linear Regression, Q : {Q_test:.3f}')
fig.subplots_adjust(hspace=0.6, wspace=0.4)
fig.savefig("fig2.png", dpi=300)
plt.show()
```

```python
# Dataset of "Price of Mazda Cars"
X = X_train_p
y = y_train

fig = plt.figure(figsize=(16,12))

alp = dict()
Q = dict()


fig = plt.figure(figsize=(16,12))

alp = dict()
Q = dict()

for k in range(1,5): #字数を指定
  # 各列のスケーリング係数
  sc = 1/np.max(X,axis=0)
  # データ行列をスケーリング
  Xs = X * sc
  # Least squares
  alpsc = la.solve(np.dot(Xs.T,Xs),np.dot(Xs.T,y))
  # リスケーリング
  alp[k] = alpsc * sc
  print('Order: {}'.format(k))
  print(' alp = ',alp[k])
  # Predicted Prices
  yhat = np.dot(X,alp[k])
  # Sum of squared errors (rediuals)
  Q[k] = np.sum((y-yhat)**2)
  print(' Q = {:.2e}'.format(Q[k]))
  # データ行列 X に列を追加
  X = np.hstack((X,(X[:,1]**(k+1)).reshape(-1,1)))
  # プロット
  ax = fig.add_subplot(2,2,k)
  ax.plot(X[:,1],y,'bo')
  x = np.linspace(min(X[:,1])-1,max(X[:,1])+1,20)
  yh = alp[k][0]*np.ones(x.shape)
  for i in range(1,k+1):
    yh += alp[k][i]*x**i
  ax.plot(x,yh,'r-')
  ax.set_title('Order: {} (Q={:.2e})'.format(k,Q[k]))

plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import scipy.linalg
```

```python
mu = [5, 10]
sigma = [[30, 20], [20, 50]] #分散共分散行列
 #ライブラリで乱数作成
x, y = np.random.multivariate_normal(mu, sigma, 10000).T

#sigmaの平方根の計算
L, V = np.linalg.eig(sigma) #固有値固有ベクトル
rL = np.diag(np.sqrt(L)) #固有値のルートを対角行列へ
A = np.dot(V, np.dot(rL, V.T))

# 平均0,分散1の正規乱数をかけるて平均値を足す。
x2, y2 = np.dot(A, np.random.randn(2, 10000))
x2 = x2 + mu[0]
y2 = y2 + mu[1]
 
k = np.polyfit(x, y, 1)  # ライブラリの乱数の1次近似
k2 = np.polyfit(x2, y2, 1)  # 自分で計算した乱数の1次近似
 
xx = np.arange(-20, 30, 1)
yy = k[0] * xx + k[1]  # 直線の式　y = ax + b
yy2 = k2[0] * xx + k2[1]
 
plt.scatter(x, y)
plt.scatter(x2, y2)
plt.plot(xx, yy, color='red')
plt.plot(xx, yy2, color='blue')
plt.xlabel("x")
plt.ylabel("y")
plt.title("2 variables gaussian plot")
plt.savefig("gaussian.png", dpi=300)
plt.show()
```

```python
import numpy as np

#Transition matrix
A = np.array([[0.6, 0.4], [0.3, 0.7]])

#Observation matrix
B = np.array([[0.6, 0.3, 0.1],[0.1, 0.4, 0.5]])

#Prior probability
pri = np.array([0.4, 0.6])

#Observation
y = np.array([0, 2, 1], dtype='int32')
tend = y.shape[0]

#Filtering distribution
alp = np.zeros((2, tend))

#Filtering
pred = pri.copy()
for i in range(tend):
  b = B[:, y[i]]
  alp[:, i] = b*pred
  alp[:, i] /= np.sum(alp[:, i])
  pred = np.dot(A.T, alp[:, i])
print(alp)
```
